```python
random.seed(42)


# Initialize Faker
fake = Faker()

# Generate patient_disease DataFrame with random disease assignment
patient_diseases = []
assigned_patients = set()
patient_disease_tracker = {}

for procedure in procedures:
    patientid = procedure["patientid"]
    procedurename = procedure["procedurename"]
    diseases = procedure_disease_mapping.get(procedurename, "Unknown Disease").split(", ")

    # Initialize patient disease set if not already done
    if patientid not in patient_disease_tracker:
        patient_disease_tracker[patientid] = set()

    # Randomly select one or more diseases from the list without duplicates
    available_diseases = [disease for disease in diseases if disease not in patient_disease_tracker[patientid]]
    if available_diseases:
        assigned_diseases = random.sample(available_diseases, k=random.randint(1, len(available_diseases)))
        for disease in assigned_diseases:
            patient_diseases.append({
                "patientid": patientid,
                "diseasename": disease
            })
            patient_disease_tracker[patientid].add(disease)
            assigned_patients.add(patientid)

# Ensure every patient has at least one disease assigned
for patientid in patients_df['patientid']:
    if patientid not in assigned_patients:
        # Assign a random disease to the patient
        procedurename = random.choice(additional_procedures)
        diseases = procedure_disease_mapping[procedurename].split(", ")
        available_diseases = [disease for disease in diseases if disease not in patient_disease_tracker.get(patientid, set())]
        if available_diseases:
            disease = random.choice(available_diseases)
        else:
            disease = random.choice(diseases)  # Fallback to assign any disease if all are already assigned

        patient_diseases.append({
            "patientid": patientid,
            "diseasename": disease
        })
        if patientid not in patient_disease_tracker:
            patient_disease_tracker[patientid] = set()
        patient_disease_tracker[patientid].add(disease)

# Create DataFrame
patient_disease_df = pd.DataFrame(patient_diseases)

patient_disease_csv_path = 'C:/Users/Admin/Downloads/patient_disease_data.csv'  # Replace this with your own local path
patient_disease_df.to_csv(patient_disease_csv_path, index=False)

print("CSV file 'patient_disease_data.csv' has been created.")
patient_disease_df.head()
```
